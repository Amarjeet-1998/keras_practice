```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_lables = []
train_samples = []
```

### Example data:


*   An experiment drug was tested on indivisuals from ages 13 to 65.
*   he trial had 2100 participants.Half were under 65 years old, half were over 65 years old
*   95% of patients 65 or older experienced side effects.
*   95% of patients under 65 experienced no side effects. 



```python
import random
for i in range(1000):
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_lables.append(0)

  random_older = randint(65,100)
  train_samples.append(random_older)
  train_lables.append(1)

for i in range(50):
  random_younger = randint(65,100)
  train_samples.append(random_younger)
  train_lables.append(1)

  random_older = randint(65,100)
  train_samples.append(random_older)
  train_lables.append(0)
```

```python
#print row data
for i in train_samples:
  print(i)
```

```python
#print row data
for i in train_lables:
  print(i)
```

```python
train_lables = np.array(train_lables)
train_samples = np.array(train_samples)
```
